<h2>Data Cleaning - Incident Ticket data</h2>
<h4>This notebook will define the data cleaning pipeline to be used in the final python script.  Based on our work in the EDA notebook, we have the following action items to perform before data is ready to be fed for training.</h4>

<ul>
    <li>Prune dataset to X and Y features</li>
    <li>Drop rows with null values in CI_CAT and CI_Subcat. </li>
    <li>One Hot Encode CI_Cat and CI_Subcat (Low cardinality, nominal)</li>
    <li>Engineer "Duration" feature as our dependent variable.</li>
    <li>Use Duration as a means of Target Encoding CI_Name and WBS (High cardinality, nominal)</li>
    <li>Normalize data types within Urgency and Time features (mixed types)</li>
    <li>Transform Time_Open to capture cyclical context of 24 hour cycle</li>
</ul>

Considerations to me made for smoothing our target encoded variables to avoid overfitting.  This will need done between splits, as we intent to cross-validate during training.
